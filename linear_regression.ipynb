```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
x = data.iloc[:, 1].values
y = data.iloc[:, 2].values
```

```python
plt.scatter(x, y, color='red', label='Data')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Prediction')
plt.legend()
plt.show()
```

```python
def loss_function(x, y, w, b):
    predictions = w * x + b
    errors = predictions - y
    return np.mean(errors**2)
```

```python
def gradient_descent(x, y, w, b, learning_rate, iterations):
    n = len(x)
    for i in range(iterations):
        predictions = w * x + b
        errors = predictions - y

        dw = (2/n) * np.dot(errors, x)
        db = (2/n) * np.sum(errors)

        w = w - learning_rate * dw
        b = b - learning_rate * db

        if i%100 == 0:
            loss = loss_function(x, y, w, b)
            print(f"Iteration {i}: Loss = {loss:.2f}, w = {w:.2f}, b = {b:.2f}")
    
    return w, b
```

```python
w, b = 0.0, 0.0
learning_rate = 0.01
iterations = 1000

w, b = gradient_descent(x, y, w, b, learning_rate, iterations)
print(f"Final parameters: w = {w:.2f}, b = {b:.2f}")
```

```python
line = w * x + b

plt.scatter(x, y, color='red', label='Data')
plt.plot(x, line, color='blue', label='Predictions')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Prediction')
plt.legend()
plt.show()
```
